## Binary Search

- Binary search is an algorithm and its input is a sorted list of elements.
  - With binary search, you keep cutting the numbers of words in half until you're left with only one word. 
  - One thing to keep in mind for a binary search is that it only works if the list is in order.

### Binary Search in Python

The binary_search function takes a sorted array and an item. If the item is in the array, the function returns its position. 

```python
def binary_search(list, item):

  low = 0
  high = len(list) - 1  # index of the last item in the list

  while (low <= high):
    mid = (low + high) // 2
    guess = list[mid]   # start guessing from the mid value. 

    # What follows after chopping in half 
    # 1) if equal: 
    if guess == item:   # we're only going to look in the bottom half of the array now.
      return mid
    # 2) if guess is greater than item
    if guess > item:
      high = mid - 1
    # 3) if guess is smaller than item
    else:               # cuts off the bottom half of the array, 
      low = mid + 1     # and we're only going to look at the top half of the array
      
  return (f'item {item} not found in the list')    # low <= high: the item doesn't exist.
```

```python
list_a = [10, 20, 30, 40]
list_b = np.arange(1, 101, 1)

def visible_binary_search(list, item):
  low = 0
  high = len(list) - 1  # index of the last item in the list

  while (low <= high):
    mid = (low + high) // 2   # Whatever the input is, just cut the list in half.
    print('mid: ', mid)
    guess = list[mid]         # find the item in list by the mid value. 
    print(f'guess: {guess}\n')
    if guess == item: # we're only going to look in the bottom half of the array now.
      print(f'guess: {guess} == item: {item}\m')
      return mid
    if guess > item:
      print(f'guess: {guess} > item: {item}')
      high = mid - 1
      print(f'high: {high} = mid {mid}\n')
    else:            #  cuts off the bottom half of the array, 
      low = mid + 1  # and we're only going to look at the top half of the array
      print(f'low {low}: mid {mid}\n')
  return (f'item {item} not found in the list')    # low <= high: the item doesn't exist.

visible_binary_search(list_b, 78)
```

### Travelling Salesperson

Run time of algorithms is expressed in big O notation. O(log n) is faster than O(n), but it gets a lot faster as the list of items you're searching grows.

- "There are a 120 permutations with 5 cities, so we'll take 120 operations to solve the problem for 5 cities. For 6 cities, it will take 720 operations, because there are 720 permutations. For 7 cities, will take 5,040 operations, and as you can see, the number of operations gets really big really quick. In general, for n items, it will take $n!\text{(n factorial)}$ operations to compute the result."

```python
 travelling_permutations = "/content/exponential growth.png"
 img = cv2.imread(travelling_permutations)
 cv2_imshow(img)
```

This is a famous problem in computer science, because its growth is awful, and some very smart people think it can't be improved. It's called the traveling salesperson problem.
